```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report
from itertools import product
from sklearn.exceptions import DataConversionWarning, ConvergenceWarning,UndefinedMetricWarning
import warnings
warnings.filterwarnings("ignore")
```

```python
X_train = pd.read_csv('../data/X_training.csv',sep=',')
y_train = pd.read_csv('../data/y_training.csv',sep=',')
X_val = pd.read_csv('../data/X_validation.csv',sep=',')
y_val = pd.read_csv('../data/y_validation.csv',sep=',')
X_test = pd.read_csv('../data/X_test.csv',sep=',')
y_test = pd.read_csv('../data/y_test.csv',sep=',')
```

```python
train = []
val = []
```

```python
model = 'KNN'

modelo = KNeighborsClassifier(n_neighbors=5)

modelo.fit(X_train, y_train)

y_train_pred = modelo.predict(X_train)
y_val_pred = modelo.predict(X_val)

accuracy_train = accuracy_score(y_train, y_train_pred)
precision_train = precision_score(y_train, y_train_pred,pos_label=1)
recall_train = recall_score(y_train, y_train_pred,pos_label=1)
f1_train = f1_score(y_train, y_train_pred,pos_label=1)

train_data = [ model , accuracy_train , precision_train , recall_train , f1_train ]
train.append(train_data)

accuracy_val = accuracy_score(y_val, y_val_pred)
precision_val = precision_score(y_val, y_val_pred,pos_label=1)
recall_val = recall_score(y_val, y_val_pred,pos_label=1)
f1_val = f1_score(y_val, y_val_pred,pos_label=1)

val_data = [model ,accuracy_val , precision_val , recall_val , f1_val ]
val.append(val_data)

k = [3,5,7,9,11,13]

best_model = []

for i in k:
    modelo = KNeighborsClassifier(n_neighbors=i)

    modelo.fit(np.concatenate((X_train,X_val),axis=0)
                            ,np.concatenate((y_train,y_val),axis=0)
              )
              
    y_test_pred = modelo.predict(X_test)

    accuracy_test = accuracy_score(y_test, y_test_pred)
    precision_test = precision_score(y_test, y_test_pred,pos_label=1)
    recall_test = recall_score(y_test, y_test_pred,pos_label=1)
    f1_test = f1_score(y_test, y_test_pred,pos_label=1)

    model_data = [model, accuracy_test, precision_test, recall_test, f1_test]
    best_model.append(model_data)
knn = pd.DataFrame(best_model,columns=['modelo' , 'ACC' , 'Precision' , 'Recall' , 'F1' ])

best_knn = knn.sort_values(by=[ 'F1' , 'ACC' , 'Precision' , 'Recall' ], ascending=[ False , False, False , False ])

best_knn = best_knn[:1]

best_knn
```

```python
model = 'DecisionTreeClassifier'

modelo = DecisionTreeClassifier()

modelo.fit(X_train, y_train)

y_train_pred = modelo.predict(X_train)
y_val_pred = modelo.predict(X_val)

accuracy_train = accuracy_score(y_train, y_train_pred)
precision_train = precision_score(y_train, y_train_pred,pos_label=1)
recall_train = recall_score(y_train, y_train_pred,pos_label=1)
f1_train = f1_score(y_train, y_train_pred,pos_label=1)

train_data = [ model , accuracy_train , precision_train , recall_train , f1_train ]
train.append(train_data)

accuracy_val = accuracy_score(y_val, y_val_pred)
precision_val = precision_score(y_val, y_val_pred,pos_label=1)
recall_val = recall_score(y_val, y_val_pred,pos_label=1)
f1_val = f1_score(y_val, y_val_pred,pos_label=1)

val_data = [model ,accuracy_val , precision_val , recall_val , f1_val ]
val.append(val_data)

max_depth_values = [None, 10, 20, 30,40,50,60]

best_model = []

for i in max_depth_values:
    modelo = DecisionTreeClassifier(max_depth=i)

    modelo.fit(np.concatenate((X_train,X_val),axis=0)
                            ,np.concatenate((y_train,y_val),axis=0)
              )
              
    y_test_pred = modelo.predict(X_test)

    accuracy_test = accuracy_score(y_test, y_test_pred)
    precision_test = precision_score(y_test, y_test_pred,pos_label=1)
    recall_test = recall_score(y_test, y_test_pred,pos_label=1)
    f1_test = f1_score(y_test, y_test_pred,pos_label=1)

    model = 'DecisionTreeClassifier'

    model_data = [model, accuracy_test, precision_test, recall_test, f1_test]
    best_model.append(model_data)
tree = pd.DataFrame(best_model,columns=['modelo' , 'ACC' , 'Precision' , 'Recall' , 'F1' ])

best_tree = tree.sort_values(by=[ 'F1' , 'ACC' , 'Precision' , 'Recall' ], ascending=[ False , False, False , False ])

best_tree = best_tree[:1]

best_tree.head()
```

```python
# Treinando o Modelo sem parametros
model = 'RandomForestClassifier'

modelo = RandomForestClassifier()

modelo.fit(X_train, y_train)

# Treinamento e validação com as métricas usando dados de treino e validação
y_train_pred = modelo.predict(X_train)
y_val_pred = modelo.predict(X_val)

accuracy_train = accuracy_score(y_train, y_train_pred)
precision_train = precision_score(y_train, y_train_pred,pos_label=1)
recall_train = recall_score(y_train, y_train_pred,pos_label=1)
f1_train = f1_score(y_train, y_train_pred,pos_label=1)

train_data = [ model , accuracy_train , precision_train , recall_train , f1_train ]
train.append(train_data)

accuracy_val = accuracy_score(y_val, y_val_pred)
precision_val = precision_score(y_val, y_val_pred,pos_label=1)
recall_val = recall_score(y_val, y_val_pred,pos_label=1)
f1_val = f1_score(y_val, y_val_pred,pos_label=1)

val_data = [model ,accuracy_val , precision_val , recall_val , f1_val ]
val.append(val_data)

#listas de Parametros
n_estimators_values = [50, 100, 150]
max_depth_values = [None, 10, 20, 30,40,50,60]

# Transformando todas as lista em uma lista de listas
combinacoes_parametros = list(product(n_estimators_values, max_depth_values))

#Lista para receber os resultados
best_model = []

# Loop para explorar todas as combinações
for n_estimators, max_depth in combinacoes_parametros:

    modelo = RandomForestClassifier(n_estimators=n_estimators,max_depth=max_depth)
    # Junção dos dados de treino e validação
    
    modelo.fit(np.concatenate((X_train,X_val),axis=0)
                            ,np.concatenate((y_train,y_val),axis=0)
              )

    # Predição para avaliar o modelo com dados de test          
    y_test_pred = modelo.predict(X_test)

    # Metricas para validação do modelo
    accuracy_test = accuracy_score(y_test, y_test_pred)
    precision_test = precision_score(y_test, y_test_pred,pos_label=1)
    recall_test = recall_score(y_test, y_test_pred,pos_label=1)
    f1_test = f1_score(y_test, y_test_pred,pos_label=1)

    model = 'RandomForestClassifier'

    model_data = [model, accuracy_test, precision_test, recall_test, f1_test]
    best_model.append(model_data)
rf = pd.DataFrame(best_model,columns=['modelo' , 'ACC' , 'Precision' , 'Recall' , 'F1' ])

# Escolha do melhor modelo com base nas metricas
best_rf = rf.sort_values(by=[ 'F1' , 'ACC' , 'Precision' , 'Recall'  ], ascending=[ False , False, False , False ])

best_rf = best_rf[:1]

best_rf
```

```python
# Treinando o Modelo sem parametros
model = 'LogisticRegression'

modelo = LogisticRegression()

modelo.fit(X_train, y_train)

# Treinamento e validação com as métricas usando dados de treino e validação
y_train_pred = modelo.predict(X_train)
y_val_pred = modelo.predict(X_val)

accuracy_train = accuracy_score(y_train, y_train_pred)
precision_train = precision_score(y_train, y_train_pred,pos_label=1)
recall_train = recall_score(y_train, y_train_pred,pos_label=1)
f1_train = f1_score(y_train, y_train_pred,pos_label=1)

train_data = [ model , accuracy_train , precision_train , recall_train , f1_train ]
train.append(train_data)

accuracy_val = accuracy_score(y_val, y_val_pred)
precision_val = precision_score(y_val, y_val_pred,pos_label=1)
recall_val = recall_score(y_val, y_val_pred,pos_label=1)
f1_val = f1_score(y_val, y_val_pred,pos_label=1)

val_data = [model ,accuracy_val , precision_val , recall_val , f1_val ]
val.append(val_data)

#listas de Parametros
valores_C = [0.1,1.0, 10.0]
valores_solver = ['newton-cg', 'lbfgs', 'liblinear']
valores_max_iter = list(range(100,400,100))

# Transformando todas as lista em uma lista de listas
combinacoes_parametros = list(product(valores_C, valores_solver, valores_max_iter))

#Lista para receber os resultados
best_model = []

# Loop para explorar todas as combinações
for C, solver, max_iter in combinacoes_parametros:

    modelo = LogisticRegression(C=C, solver=solver, max_iter=max_iter)
    # Junção dos dados de treino e validação
    
    modelo.fit(np.concatenate((X_train,X_val),axis=0)
                            ,np.concatenate((y_train,y_val),axis=0)
              )

    # Predição para avaliar o modelo com dados de test          
    y_test_pred = modelo.predict(X_test)

    # Metricas para validação do modelo
    accuracy_test = accuracy_score(y_test, y_test_pred)
    precision_test = precision_score(y_test, y_test_pred,pos_label=1)
    recall_test = recall_score(y_test, y_test_pred,pos_label=1)
    f1_test = f1_score(y_test, y_test_pred,pos_label=1)

    model = 'LogisticRegression'

    model_data = [model, accuracy_test, precision_test, recall_test, f1_test]
    best_model.append(model_data)
log = pd.DataFrame(best_model,columns=['modelo' , 'ACC' , 'Precision' , 'Recall' , 'F1' ])

# Escolha do melhor modelo com base nas metricas
best_log = log.sort_values(by=[ 'F1' , 'ACC' , 'Precision' , 'Recall'  ], ascending=[ False , False, False , False ])

best_log = best_log[:1]

best_log.head()
```
