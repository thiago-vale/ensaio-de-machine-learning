```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report
from itertools import product
from sklearn.exceptions import DataConversionWarning, ConvergenceWarning,UndefinedMetricWarning
import warnings
warnings.filterwarnings("ignore", category=DataConversionWarning)
warnings.filterwarnings("ignore", category=ConvergenceWarning)
warnings.filterwarnings("ignore", category=UndefinedMetricWarning)
```

```python
X_train = pd.read_csv('../data/X_training.csv',sep=',')
y_train = pd.read_csv('../data/y_training.csv',sep=',')
X_val = pd.read_csv('../data/X_validation.csv',sep=',')
y_val = pd.read_csv('../data/y_validation.csv',sep=',')
X_test = pd.read_csv('../data/X_test.csv',sep=',')
y_test = pd.read_csv('../data/y_test.csv',sep=',')
```

```python

modelo = 'LogisticRegression'

valores_C = [0.1,1.0, 10.0]
valores_solver = ['lbfgs', 'liblinear']
valores_max_iter = list(range(100,500,100))

combinacoes_parametros = list(product(valores_C, valores_solver, valores_max_iter))

for C, solver, max_iter in combinacoes_parametros:
    modelo = LogisticRegression(C=C, solver=solver, max_iter=max_iter)
    modelo.fit(X_train, y_train)

    y_train_pred = modelo.predict(X_train)
    y_val_pred = modelo.predict(X_val)
    y_test_pred = modelo.predict(X_test)

        
    accuracy_train = accuracy_score(y_train, y_train_pred)  # Substitua y_test pelas suas labels de teste
    precision_train = precision_score(y_train, y_train_pred,zero_division=1)
    recall_train = recall_score(y_train, y_train_pred)
    f1_train = f1_score(y_train, y_train_pred)
    class_train = classification_report(y_train, y_train_pred)

    accuracy_val = accuracy_score(y_val, y_val_pred)  # Substitua y_test pelas suas labels de teste
    precision_val = precision_score(y_val, y_val_pred,zero_division=1)
    recall_val = recall_score(y_val, y_val_pred)
    f1_val = f1_score(y_val, y_val_pred)
    class_val = classification_report(y_val, y_val_pred)

    accuracy_test = accuracy_score(y_test, y_test_pred)  # Substitua y_test pelas suas labels de teste
    precision_test = precision_score(y_test, y_test_pred,zero_division=1)
    recall_test = recall_score(y_test, y_test_pred)
    f1_test = f1_score(y_test, y_test_pred)
    class_teste = classification_report(y_test, y_test_pred)

    print('params')
    print(f'params: C={C}, solver={solver}, max_iter={max_iter}')
    print('train')
    print(class_train)
    print('val')
    print(class_val)
    print('test')
    print(class_teste)

    #print('Train')
    #print(f'params: C={C}, solver={solver}, max_iter={max_iter}')
    #print(f'Accuracy Train: {accuracy_train:.3f}')
    #print(f'Precision Train: {precision_train:.3f}')
    #print(f'Recall Train: {recall_train:.3f}')
    #print(f'F1-Score Train: {f1_train:.3f}')
    #print('Val')
    #print(f'Accuracy Val: {accuracy_val:.3f}')
    #print(f'Precision Val: {precision_val:.3f}')
    #print(f'Recall Val: {recall_val:.3f}')
    #print(f'F1-Score Val: {f1_val:.3f}')
    #print('Test')
    #print(f'Accuracy Test: {accuracy_test:.3f}')
    #print(f'Precision Test: {precision_test:.3f}')
    #print(f'Recall Test: {recall_test:.3f}')
    #print(f'F1-Score Test: {f1_test:.3f}')
    #print('-' * 50)
```
